```python
from transformers import pipeline, BlipProcessor, BlipForConditionalGeneration
from diffusers import StableDiffusionPipeline
from PIL import Image
import torch
import matplotlib.pyplot as plt

text_generator = pipeline("text-generation", model="gpt2")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

device = "cpu"
sd_model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4").to(device)

processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

def generate_text(title, max_length=100):
    """Generate a textual description based on the input title."""
    text = text_generator(title, max_length=max_length, num_return_sequences=1)
    return text[0]['generated_text']

def expand_text(text, max_length=200):
    """Expand the given text by generating more content."""
    expanded_text = text_generator(text, max_length=max_length, num_return_sequences=1)
    return expanded_text[0]['generated_text']

def summarize_text(text, max_length=50):
    """Summarize the generated text."""
    summary = summarizer(text, max_length=max_length, min_length=25, do_sample=False)
    return summary[0]['summary_text']

def generate_image(prompt):
    """Generate an image based on the input prompt using Stable Diffusion."""
    with torch.autocast("cpu"):
        image = sd_model(prompt).images[0]
    return image

def describe_image(image):
    """Generate a description for the given image using the BLIP model."""
    inputs = processor(image, return_tensors="pt").to(device)
    out = model.generate(**inputs, max_new_tokens=20)
    description = processor.decode(out[0], skip_special_tokens=True)
    return description

def display_image(image):
    """Display the generated image using matplotlib in Kaggle."""
    plt.imshow(image)
    plt.axis("off")
    plt.show()

def main():
    title = input("Enter a topic title: ")

    initial_text = generate_text(title)
    print(f"\nGenerated Text:\n{initial_text}")

    expand = input("Would you like to expand the text? (yes/no): ")
    if expand.lower() == "yes":
        expanded_text = expand_text(initial_text)
        print(f"\nExpanded Text:\n{expanded_text}")
    else:
        expanded_text = initial_text

    print("\nGenerating image based on the title...")
    image = generate_image(title)

    display_image(image)

    summary = summarize_text(expanded_text)
    print(f"\nSummary of the Text:\n{summary}")

    image_description = describe_image(image)
    print(f"\nImage Description:\n{image_description}")

if __name__ == "__main__":
    main()
```
